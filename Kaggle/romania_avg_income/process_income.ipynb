```python
import pandas as pd
import numpy as np
```

```python
dtypes={'January': str, 'February': str, 'March': str, 'April': str, 'May': str, 'June': str,
        'July': str, 'August': str, 'September': str, 'October': str, 'November': str, 'December': str}
```

```python
net = pd.read_excel('Romania earnings - since 1991.xlsx', sheet_name="net", dtype=dtypes)
gross = pd.read_excel('Romania earnings - since 1991.xlsx', sheet_name="gross", dtype=dtypes)
```

```python
net.head()
```

```python
gross.head()
```

```python
new_columns = ['Year', 'January', 'February', 'March', 'April', 'May', 'June',
        'July', 'August', 'September', 'October', 'November', 'December']
net.columns = new_columns
gross.columns = new_columns
```

```python
net_df = net.melt(id_vars=["Year"], var_name="Month", value_name="Earnings")
```

```python
gross_df = gross.melt(id_vars=["Year"], var_name="Month", value_name="Earnings")
```

```python
net_df.head()
```

```python
gross_df.head()
```

```python
net_df.info()
```

```python
gross_df.info()
```

```python
net_df['Earnings'] = net_df['Earnings'].apply(lambda x: str(x))
net_df['Earnings'] = net_df['Earnings'].apply(lambda x: x.replace(" ", ""))
net_df['Earnings'] = net_df['Earnings'].apply(lambda x: x.replace(".", ""))

gross_df['Earnings'] = gross_df['Earnings'].apply(lambda x: str(x))
gross_df['Earnings'] = gross_df['Earnings'].apply(lambda x: x.replace(" ", ""))
gross_df['Earnings'] = gross_df['Earnings'].apply(lambda x: x.replace(".", ""))
                                                                                                                                          
```

```python
net_df.to_csv("net_monthly_average_earnings.csv", index=False)
```

```python
gross_df.to_csv("gross_monthly_average_earnings.csv", index=False)
```
