```python
dna_df = pd.read_csv('doctoral_boards_defence_national_academy.csv')
nia_df = pd.read_csv('doctoral_boards_national_intelligence_academy.csv')
pa_df = pd.read_csv('doctoral_boards_police_academy.csv')
```

```python
dna_df.head()
```

```python
dna_df.Role.value_counts()
```

```python
dna_df['Title'] = None
```

```python
dna_df.columns = ['Year', 'No', 'Candidate', 'Board_Member', 'Role', 'Board', 'Name', 'Title']
dna_df = dna_df[['No', 'Year', 'Candidate','Title', 'Board_Member', 'Role', 'Name']]
dna_df.head()
```

```python
dna_df['Role'] = dna_df['Role'].apply(lambda x: x.replace("Presedinte", "Chairman"))
dna_df['Role'] = dna_df['Role'].apply(lambda x: x.replace("Coordonator", "Advisor"))
dna_df['Role'] = dna_df['Role'].apply(lambda x: x.replace("Referent", "Reviewer"))
```

```python
nia_df.columns = ['No', 'Candidate', 'Title', 'Advisor', 'Board', 'Year', 'Board_Member', 'Role', 'Name']
nia_df = nia_df[['No', 'Year', 'Candidate','Title', 'Board_Member', 'Role', 'Name']]
nia_df.head()
```

```python
nia_df['Role'] = nia_df['Role'].apply(lambda x: x.replace("Presedinte", "Chairman"))
nia_df['Role'] = nia_df['Role'].apply(lambda x: x.replace("Coordonator", "Advisor"))
nia_df['Role'] = nia_df['Role'].apply(lambda x: x.replace("Referent", "Reviewer"))
```

```python
pa_df.columns = ['No', 'Title', 'Year', 'Candidate', 'Board', 'Board_Member', 'Role', 'Name']
pa_df = pa_df[['No', 'Year', 'Candidate','Title', 'Board_Member', 'Role', 'Name']]
pa_df.head()
```

```python
pa_df['Role'] = pa_df['Role'].apply(lambda x: x.replace("Presedinte", "Chairman"))
pa_df['Role'] = pa_df['Role'].apply(lambda x: x.replace("Coordonator", "Advisor"))
pa_df['Role'] = pa_df['Role'].apply(lambda x: x.replace("Referent", "Reviewer"))
```
